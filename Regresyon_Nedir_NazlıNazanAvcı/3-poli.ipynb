```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('salaries.csv')
X=dataset.iloc[:, 1:-1].values
y=dataset.iloc[:, -1].values
```

```python
from sklearn.linear_model import LinearRegression
linearREG=LinearRegression()
linearREG.fit(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
polyREG=PolynomialFeatures(degree=4)
X_poly=polyREG.fit_transform(X)

linearREG_2=LinearRegression()
linearREG_2.fit(X_poly, y)
```

```python
#X_poly
```

```python
plt.scatter(X, y, color="red")
plt.plot(X, linearREG.predict(X), color="blue")
plt.show()
```

burda linear Regresyon uygun bir model değil. Burda poly Regresyon kullanabiliriz.

```python
plt.scatter(X, y, color="red")
plt.plot(X,linearREG_2.predict(X_poly), color="blue")
plt.show()
```

Bu da tam olmadı. Yukardaki dereceyi artırıralım. Ne kadar artırırsak o kadar cok performans ve vakit gider

```python
plt.scatter(X, y, color="red")
plt.plot(X,linearREG_2.predict(X_poly), color="blue")
plt.show()
```

```python
linearREG.predict([[8.5]])
```

```python
linearREG_2.predict(polyREG.fit_transform([[8.5]]))
```

```python
X_grid=np.arange(min(X), max(X), 0.1)
X_grid=X_grid.reshape(len(X_grid), 1)
```

```python
plt.scatter(X, y, color="red")
plt.plot(X_grid,linearREG_2.predict(polyREG.fit_transform(X_grid)), color="blue")
plt.show()
```
